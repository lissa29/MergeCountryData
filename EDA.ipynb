```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('unemployment analysis.csv' , sep=",")
```

```python
df.dtypes
```

```python
df.describe()
```

```python
dfnew = pd.DataFrame()
for y in df.columns[ 2: ]:
    dftemp = df.loc[ : , ["Country Name","Country Code",y] ]
    dftemp["Year"] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Unemployment value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Unemployment value'] ]
    
    dfnew = pd.concat([dfnew, dftemp])
```

```python
dfh = pd.read_csv('Human_life_Expectancy.csv' , sep=",")
```

```python
dfh
```

```python
dfhnew = pd.DataFrame()
for a in dfh.columns[ 5: ]:
    dftempo = dfh.loc[ : , ["Country", "Country Code","Level","Region", a] ]
    dftempo["Year"] = a
    dftempo.columns=['Country', 'Country Code','Level','Region', 'Human Life Expectancy value','Year'] 
    dftempo = dftempo.loc[ : , ['Country', 'Country Code','Level','Region', 'Year','Human Life Expectancy value'] ]
    
    
    dfhnew = pd.concat([dfhnew,dftempo])
```

```python
all_data= pd.merge(dfnew, dfhnew, on = 'Country Code', how = "left")
```
